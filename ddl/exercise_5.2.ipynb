```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans #The k-means algorithm
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
sns.scatterplot(x = "PetalLengthCm", y = "SepalLengthCm", style = "Species", data = df)
```

There is one very obvious cluster to be seen, the Iris-setosa has both a shorter sepal and petal than the other two species. This causes an obvious cluster to show. <br>
Where the other two species slightly combine into one larger and less obvious cluster, it is clear that overall the Iris-virinica has both longer petals and longer sepals than the Iris-versicolor. It is definitely less obvious than the first cluster, though.

#### opracht 
Using k-means clustering with 3 clusters (using only petal length and sepal length as X variables), create a new variable cluster and use this to make another scatterplot, using color for cluster.


```python
km = KMeans(n_clusters=3)
X = df[['SepalLengthCm', 'PetalLengthCm']]
km = km.fit(X)
df['Cluster'] = km.predict(X)
df.head()
```

```python
sns.scatterplot(x = "PetalLengthCm", y = "SepalLengthCm", hue = 'Cluster', style = "Species", data = df)
```

K-means performs relatively okay, however, there is one obvious outlier that doesn't make too much sense (the cluster 0 point at 3 Petal Length Cm). It seems to be located much closer to the cluster 2 when compared to the cluster 0. <br>
It's also not completely clear where the line is drawn when looking at cluster 1 and cluster 2, since many datapoints are so close to each other. 

```python
km = KMeans(n_clusters=2)
X = df[['SepalLengthCm', 'PetalLengthCm']]
km = km.fit(X)
df['Cluster 2'] = km.predict(X)
df.head()
```

```python
sns.scatterplot(x = "PetalLengthCm", y = "SepalLengthCm", hue = 'Cluster 2', style = "Species", data = df)
```

```python
km = KMeans(n_clusters=4)
X = df[['SepalLengthCm', 'PetalLengthCm']]
km = km.fit(X)
df['Cluster 4'] = km.predict(X)
df.head()
```

```python
sns.scatterplot(x = "PetalLengthCm", y = "SepalLengthCm", hue = 'Cluster 4', style = "Species", data = df)
```

Where the natural amount of clusters seems like three (for the three different species), I personally feel like the 4 different clusters makes more sense. Drawing the line (that made the three clusters difficult to interpret) in the one bigger cluster now seems more obvious.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_movies = pd.read_csv('movies.csv')
df_movies.head()
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
df_combined = pd.merge(df_ratings, df_movies, on = 'movieId')
df_combined.head()
```

```python
pivot = pd.pivot_table(data = df_combined, index='userId', columns='title', values='rating')
pivot.head()

#aggfunc = 'sum', or 'mean', or 'count' can be added to this as well.
```

```python
sns.distplot(pivot["Harry Potter and the Sorcerer's Stone (a.k.a. Harry Potter and the Philosopher's Stone) (2001)"].dropna(), kde=False, bins=10)
```

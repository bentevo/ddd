```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df_movies = pd.read_csv('movies.csv')
df_ratings = pd.read_csv('ratings.csv')
```

```python
df_combined = pd.merge(df_ratings, df_movies, on = 'movieId')
df_combined.head()
```

```python
pivot = pd.pivot_table(data = df_combined, index='userId', columns='title', values='rating')
pivot.head()
```

```python
pivot = pivot.reindex(pivot.count().sort_values(ascending=False).index, axis=1)
pivot.head()
```

```python
pivot_filled = pivot.fillna(0) #fill the NaN with the mean of each column
pivot_filled.head()
```

```python
km = KMeans(n_clusters=5)
X = pivot_filled.loc[:,'Forrest Gump (1994)':"'71 (2014)"]
km = km.fit(X)
pivot_filled['Cluster'] = km.predict(X)
pivot_filled.head()
```

```python
pivot_filled['Cluster'].value_counts()
```

```python
pivot_filled = pd.pivot_table(data = pivot_filled, index = 'Cluster')

```

```python
pivot_filled = pivot_filled.reindex(pivot.count().sort_values(ascending=False).index, axis=1)
pivot_filled.head()
```

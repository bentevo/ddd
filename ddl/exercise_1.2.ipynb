```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('class_data.csv')
df
```

```python
df[df['instagram_followers']<500]
```

```python
df2 =df[df['instagram_followers']<500]
```

```python
sns.boxplot(df2['instagram_followers'])
plt.title('Instagram Followers')
plt.xlabel('Amount of followers')
plt.show()
```

```python
linkedincon = df['linkedin_con'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(linkedincon, kde=False) 
plt.title('LinkedIn Connections')
plt.xlabel('Amount of connections')
plt.show()
```

```python
df_app = df[df['apps_phone'] > 0] #Subsetting the dataframe to include only paid apps
df_app = df[df['apps_phone'] < 1000] #Subsetting the dataframe to include only paid apps
sns.violinplot(df_app['apps_phone'])
sns.swarmplot(df_app['apps_phone'], color='purple') #default color doesn't stand out
plt.title('Number of applications on phone')
plt.ylabel('Number of students')
plt.xlabel('Amount of applications')
plt.show()
```

```python
mean_size = df['texts_yesterday'].mean()
median_size = df['texts_yesterday'].median()

print(f'The median filesize is: {median_size} texts')
print(f'The mean filesize is: {mean_size} texts')

df['texts_yesterday'].name = 'Texts Yesterday' #Labeling the variable for plotting
textsyesterday = df['texts_yesterday'].dropna() 

sns.distplot(textsyesterday, kde=False) #I need to drop the NA's, otherwise Seaborn won't plot
plt.title('Number of texts sent yesterday')
plt.axvline(median_size, 0, 100, color='green', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_size, 0, 100, color='blue', label='mean') 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

```python
phonesowned = df['phones_owned'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(phonesowned) 
plt.title('Number of phones owned in lifetime')
plt.xlabel('Number of phones')
plt.ylabel('% of respondents')
plt.show()
```

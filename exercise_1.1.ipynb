```python
import pandas as pd

#seperate on ;
df1 = pd.read_csv('steps_data_1.csv', sep=';')
df1.head()
```

```python
df2 = pd.read_csv('survey_data_1.csv')
df2.head()
```

```python
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head()
```

```python
df['waist'].plot()
```

```python
df['waist'].plot(kind = 'hist', bins = 20)
```

```python
def remove_extra_waist(x):
    if(x > 130.0): 
        return float('NaN')
    if(x < 45.0): 
        return float('NaN')
    else: 
        return x

df['waist'] = df['waist'].apply(remove_extra_waist)
```

```python
df['waist'].plot(kind = 'hist')
```
